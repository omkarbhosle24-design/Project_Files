```python
# 5.1 Analyze the peak number of daily new cases in Germany, France, and Italy. 
# Which country experienced the highest single-day surge, and when did it occur?

target_countries = ['Germany' , 'France' , 'Italy']
country_date_total = (
    merged.loc[merged['Country/Region'].isin(target_countries)]
    .groupby(['Country/Region','Date'])['Confirmed'].sum().reset_index()
    .sort_values(['Country/Region','Date'])
)
country_date_total
```

```python
# Cumulative Confirmed Cases Convert -> Daily New Confirmed Cases [.diff]
country_date_total['DailyNew'] = (
    country_date_total.groupby('Country/Region')['Confirmed'].diff()
)
country_date_total
```

```python
# Identifying the missing Values
country_date_total[country_date_total['DailyNew'].isna()]
```

```python
country_date_total['DailyNew'] = (
    country_date_total.groupby('Country/Region')['Confirmed'].diff()
    .fillna(country_date_total['Confirmed'])
)
country_date_total
```

```python
# Identifying the missing Values
country_date_total[country_date_total['DailyNew'].isna()]
```

```python
# Pivot Table
daily_new_matrix = (
    country_date_total.pivot(
        index = 'Date',
        columns = 'Country/Region',
        values = 'DailyNew'
    ).sort_index()
)
daily_new_matrix
```

```python
daily_new_matrix['France'].nlargest(5)
```

```python
daily_new_matrix['Germany'].nlargest(5)
```

```python
daily_new_matrix['Italy'].nlargest(5)
```

```python
daily_new_by_country = {}

# target_countries = ['Germany' , 'France' , 'Italy']

for country in target_countries:
    if country in daily_new_matrix.columns:
        s = daily_new_matrix[country]
        daily_new_by_country[country] = s

daily_new_by_country
```

```python
daily_new_matrix['Germany']
```

```python
daily_new_matrix['France']
```

```python
# Peak {Country : (max_value , date} per date
peaks = {}
for country , series in daily_new_by_country.items():
    max_value = int(series.max())
    max_index = series.idxmax()
    peaks[country] = (max_value , max_index)

peaks
```

```python
# Visualizations
plt.figure(figsize = (14,8) , facecolor = '#E8E8E8')
for country , series in daily_new_by_country.items():
    plt.plot(series.index , series.values , label = country)

plt.title('Daily New Confirmed Cases For Germany, France , Italy.' , color = 'red')
plt.xlabel('Date')
plt.ylabel('Confirmed Cases')
plt.ylim(-200000 , 200000)
plt.legend()
plt.show()
```

```python
# 5.2 Compare the recovery rates (recoveries/confirmed cases) between Canada and Australia 
# as of December 31, 2020. Which country showed better management of the pandemic according to this metric?

countries = ['Canada' , 'Australia']
target_date = pd.Timestamp('2020-12-31')

# .loc[rows,cols]
recovery_rate = (
    merged.loc[(merged['Country/Region'].isin(countries)) & (merged['Date'] == (target_date)),
    ['Country/Region' , 'Recovered' , 'Confirmed']]
)
recovery_rate
```

```python
recovery_rate['RecoveryRate%'] = (recovery_rate['Recovered'] / recovery_rate['Confirmed'] * 100).round(2)
recovery_rate
```

```python
# Visualization [Column Chart] [BarPlot]
plt.figure(figsize = (7,4) , facecolor = '#E8E8E8')
bars = plt.bar(recovery_rate['Country/Region'] , recovery_rate['RecoveryRate%'],
              color = ['#4D59D6' , '#E9724D'])
plt.title('Recovery Rate b/w Australia & Canada' , color = 'red')
plt.xlabel('Country')
plt.ylabel('Recovery Rate %')
plt.ylim(0,100)
plt.bar_label(bars , fmt = '%.2f%%')
plt.show()
```

```python
# 5.3 What is the distribution of death rates (deaths/confirmed cases) among provinces in Canada? 
# Identify the province with the highest and lowest death rate as of the latest data point.
confirmed_long[confirmed_long['Country/Region'] == 'Canada']
```

```python
deaths_long[deaths_long['Country/Region'] == 'Canada']
```

```python
canada_confirmed_cases = (
    confirmed_long[confirmed_long['Country/Region'] == 'Canada']
    .sort_values('Date')
    .groupby(['Province/State']).tail(1)[['Province/State','Confirmed']]
    .set_index(['Province/State'])
)
canada_confirmed_cases
```

```python
canada_deaths_cases = (
    deaths_long[deaths_long['Country/Region'] == 'Canada']
    .sort_values('Date')
    .groupby(['Province/State']).tail(1)[['Province/State','Deaths']]
    .set_index(['Province/State'])
)
canada_deaths_cases
```

```python
# .join Concatenate Columns [V-LookUp] [Related Function in DAX]
canada = canada_deaths_cases.join(canada_confirmed_cases , how = 'inner')
canada
```

```python
# Death Rate [Deaths / Confirmed] %
canada['DeathRate'] = (canada['Deaths'] / canada['Confirmed'] * 100).round(2)
canada
```

```python
# Death Rate [Deaths / Confirmed] %
import numpy as np
canada['DeathRate'] = (canada['Deaths'] / canada['Confirmed'] * 100).round(2).replace([np.inf],0)
canada
```

```python
# Visualization [Bar Plot]
plt.figure(figsize = (14,8) , facecolor = '#E8E8E8')
canada['DeathRate'].sort_values(ascending = False).plot(kind = 'bar') # High to Low
plt.title('Covid19 Death Rates Comparison on all Province of Canada' , color = 'red')
plt.xlabel('Province/State')
plt.ylabel('Death Rates')
plt.xticks(rotation=45)
plt.show()
```
